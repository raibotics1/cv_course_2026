```python
import os
import random
from pathlib import Path
from dataclasses import dataclass, field

import numpy as np
import torch
import MinkowskiEngine as ME

@dataclass(frozen=True)
class ModelConfig:
    """Неизменяемая конфигурация гиперпараметров."""
    voxel_size: float = 0.05
    epochs: int = 5
    batch_size: int = 4
    lr: float = 1e-3
    weight_decay: float = 1e-4
    seed: int = 777
    device: torch.device = field(
        default_factory=lambda: torch.device("cuda" if torch.cuda.is_available() else "cpu")
    )
    # Параметры данных
    train_steps: int = 250
    val_steps: int = 80
    num_workers: int = 2
    checkpoint_path: str = "minkfcnn_modelnet40.pth"

class EnvironmentManager:
    """Отвечает за воспроизводимость и подготовку железа."""
    def __init__(self, config: ModelConfig):
        self.config = config
        self._setup_reproducibility()

    def _setup_reproducibility(self):
        seed = self.config.seed
        random.seed(seed)
        np.random.seed(seed)
        torch.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        # Для максимальной детерминированности (может замедлить обучение)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
        
    def announce(self):
        print(f"🚀 Инициализация на устройстве: {self.config.device.type.upper()}")
        if torch.cuda.is_available():
            print(f"🔥 GPU: {torch.cuda.get_device_name(0)}")
```

```python
class PointCloudTransform:
    """Базовый интерфейс для аугментаций."""
    def __call__(self, coords: np.ndarray) -> np.ndarray:
        return coords

class RandomScale(PointCloudTransform):
    def __init__(self, low=0.9, high=1.1, p=0.9):
        self.range, self.p = (low, high), p
    def __call__(self, coords):
        if random.random() < self.p:
            coords *= np.random.uniform(*self.range, size=[1, 3])
        return coords

class RandomTranslation(PointCloudTransform):
    def __init__(self, magnitude=0.2, p=0.9):
        self.mag, self.p = magnitude, p
    def __call__(self, coords):
        if self.mag > 0 and random.random() < self.p:
            coords += np.random.uniform(-self.mag, self.mag, size=[1, 3])
        return coords

class RandomJitter(PointCloudTransform):
    def __init__(self, sigma=0.025, clip=0.05, p=0.7):
        self.sigma, self.clip, self.p = sigma, clip, p
    def __call__(self, coords):
        if random.random() < self.p:
            jitter = np.clip(self.sigma * (np.random.rand(len(coords), 3) - 0.5), -self.clip, self.clip)
            coords += jitter
        return coords

class Compose:
    """Последовательное применение трансформаций."""
    def __init__(self, transforms):
        self.transforms = transforms
    def __call__(self, x):
        for t in self.transforms: x = t(x)
        return x
```

```python
class DatasetDownloader:
    @staticmethod
    def fetch_modelnet40(target_dir="modelnet40_ply_hdf5_2048"):
        root = Path(target_dir)
        if root.exists(): return
        
        url = "https://huggingface.co/datasets/Msun/modelnet40/resolve/main/modelnet40_ply_hdf5_2048.zip"
        zip_path = Path("modelnet.zip")
        
        print(f"📥 Downloading dataset to {zip_path}...")
        try:
            urllib.request.urlretrieve(url, zip_path)
            with zipfile.ZipFile(zip_path) as zf:
                zf.extractall(".")
            print("✅ Extraction complete.")
        finally:
            if zip_path.exists(): zip_path.unlink()

class ModelNetDataset(Dataset):
    def __init__(self, phase, transform=None, num_points=2048):
        DatasetDownloader.fetch_modelnet40()
        self.phase = "test" if phase in ["val", "test"] else "train"
        self.transform = transform
        self.num_points = num_points
        self.coords, self.labels = self._load_h5_data()

    def _load_h5_data(self):
        path_pattern = f"modelnet40_ply_hdf5_2048/ply_data_{self.phase}*.h5"
        files = glob.glob(path_pattern)
        
        all_data, all_labels = [], []
        for f_path in files:
            with h5py.File(f_path, 'r') as f:
                all_data.append(f["data"][:])
                all_labels.append(f["label"][:])
        
        return np.concatenate(all_data).astype(np.float32), \
               np.concatenate(all_labels).astype(np.int64)

    def __getitem__(self, idx):
        points = self.coords[idx]
        
        if self.phase == "train":
            np.random.shuffle(points)
            
        points = points[:self.num_points]
        if self.transform:
            points = self.transform(points)
            
        points_tensor = torch.from_numpy(points)
        return {
            "coordinates": points_tensor,
            "features": points_tensor, # Использование координат как фич
            "label": torch.from_numpy(self.labels[idx])
        }

    def __len__(self):
        return len(self.coords)
```

```python
def make_res_block(in_c, out_c, kernel=3, stride=1, D=3):
    """Универсальный сверточный блок: Conv -> BN -> Act."""
    return nn.Sequential(
        ME.MinkowskiConvolution(in_c, out_c, kernel_size=kernel, stride=stride, dimension=D),
        ME.MinkowskiBatchNorm(out_c),
        ME.MinkowskiLeakyReLU()
    )

def make_mlp(in_c, out_c):
    """Блок полносвязного слоя для разреженных данных."""
    return nn.Sequential(
        ME.MinkowskiLinear(in_c, out_c, bias=False),
        ME.MinkowskiBatchNorm(out_c),
        ME.MinkowskiLeakyReLU()
    )


class MinkowskiFCNN(ME.MinkowskiNetwork):
    def __init__(self, in_channel, out_channel, embedding_dim=256, channels=(16, 24, 32, 48, 64), D=3):
        super().__init__(D)
        
        # 1. Входной лифт (Embedding)
        self.input_layer = make_mlp(in_channel, channels[0])
        
        # 2. Пирамида сверток
        self.layers = nn.ModuleList([
            make_res_block(channels[i], channels[i+1], stride=1 if i==0 else 2, D=D)
            for i in range(len(channels) - 1)
        ])
        
        self.pool = ME.MinkowskiMaxPooling(kernel_size=3, stride=2, dimension=D)
        
        # 3. Агрегация признаков (Feature Fusion)
        # Суммируем каналы всех промежуточных слоев (кроме первого)
        concat_channels = sum(channels[1:])
        
        self.bottleneck = nn.Sequential(
            make_res_block(concat_channels, embedding_dim // 4, stride=2, D=D),
            make_res_block(embedding_dim // 4, embedding_dim // 2, stride=2, D=D),
            make_res_block(embedding_dim // 2, embedding_dim, stride=2, D=D),
        )
        
        # 4. Глобальные пулинги
        self.max_pool = ME.MinkowskiGlobalMaxPooling()
        self.avg_pool = ME.MinkowskiGlobalAvgPooling()
        
        # 5. Классификатор
        self.classifier = nn.Sequential(
            make_mlp(embedding_dim * 2, 256),
            ME.MinkowskiDropout(),
            ME.MinkowskiLinear(256, out_channel, bias=True)
        )
        
        self.apply(self._init_weights)

    def _init_weights(self, m):
        if isinstance(m, ME.MinkowskiConvolution):
            ME.utils.kaiming_normal_(m.kernel, mode="fan_out", nonlinearity="relu")
        elif isinstance(m, ME.MinkowskiBatchNorm):
            nn.init.constant_(m.bn.weight, 1)
            nn.init.constant_(m.bn.bias, 0)

    def forward(self, x: ME.SparseTensor):
        x = self.input_layer(x)
        
        intermediates = []
        curr = x
        
        # Проходим по слоям и собираем признаки для конкатенации
        for i, layer in enumerate(self.layers):
            curr = layer(curr)
            if i > 0: # Пропускаем первый слой для соответствия вашей логике concat
                curr = self.pool(curr)
            intermediates.append(curr)
            
        # Feature Fusion через слайсинг к исходному разрешению или между собой
        # Здесь мы приводим все тензоры к одной пространственной сетке (координатам x)
        features = ME.cat(*[feat.slice(x) for feat in intermediates])
        
        # Сжатие и глобальный пулинг
        latent = self.bottleneck(features)
        
        global_feat = ME.cat(self.max_pool(latent), self.avg_pool(latent))
        
        return self.classifier(global_feat).F # Возвращаем только фичи (Dense Tensor)

# Инициализация
model = MinkowskiFCNN(in_channel=3, out_channel=40).to(device)
```

```python
class PointCloudProcessor:
    """Утилиты для подготовки данных перед подачей в сеть."""
    @staticmethod
    def to_sparse_field(batch, voxel_size, device):
        coords = batch["coordinates"].clone()
        # Квантование координат (переход к воксельной сетке)
        coords[:, 1:] /= voxel_size 
        return ME.TensorField(
            coordinates=coords, 
            features=batch["features"], 
            device=device
        )

def get_criterion(smoothing=0.1):
    """Используем встроенный функционал для сглаживания меток."""
    return nn.CrossEntropyLoss(label_smoothing=smoothing)
```

```python
import time
from datetime import datetime

class ModelNetTrainer:
    def __init__(self, model, config, train_loader, val_loader):
        self.model = model
        self.cfg = config
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.optimizer = optim.AdamW(model.parameters(), lr=config.lr, weight_decay=config.weight_decay)
        self.scheduler = optim.lr_scheduler.CosineAnnealingLR(
            self.optimizer, T_max=config.epochs * config.train_batches_per_epoch
        )
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
        self.best_acc = 0.0
        self.history = []

    def _step(self, batch, is_train=True):
        coords, features, labels = batch
        labels = labels.to(self.cfg.device).view(-1)
        
        # 1. Подготовка координат (4 колонки: Batch, X, Y, Z)
        coords_norm = coords.clone().float()
        coords_norm[:, 1:] /= self.cfg.voxel_size
        
        # 2. Подготовка признаков (3 колонки: X, Y, Z)
        # Важно: берем только последние 3 колонки, если в признаках затесался Batch Index
        if features.shape[1] == 4:
            features = features[:, 1:] 
        
        features = features.to(self.cfg.device).float()
        
        # 3. Создание TensorField
        inputs = ME.TensorField(
            features=features,
            coordinates=coords_norm,
            device=self.cfg.device
        )
        
        logits = self.model(inputs)
        
        # Если модель возвращает SparseTensor или TensorField, берем .F (фичи)
        if hasattr(logits, 'F'):
            logits = logits.F
    
        loss = self.criterion(logits, labels)
        
        if is_train:
            self.optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), 10.0)
            self.optimizer.step()
            self.scheduler.step()
                
        return loss.item(), (logits.argmax(1) == labels).sum().item(), len(labels)

    def run_epoch(self, epoch):
        self.model.train()
        m = {"loss": 0, "correct": 0, "total": 0}
        train_iter = iter(self.train_loader)
        start_time = time.time()

        print(f"\n{'='*20} ЭПОХА {epoch}/{self.cfg.epochs} {'='*20}")
        
        for i in range(1, self.cfg.train_batches_per_epoch + 1):
            try: batch = next(train_iter)
            except StopIteration: 
                train_iter = iter(self.train_loader)
                batch = next(train_iter)
            
            l, c, t = self._step(batch)
            m["loss"] += l * t; m["correct"] += c; m["total"] += t
            
            if i % self.cfg.stat_freq == 0:
                acc = m["correct"] / m["total"]
                elapsed = time.time() - start_time
                print(f"  [Batch {i:3d}] ⚡ Loss: {l:.4f} | Acc: {acc:.2%} | Time: {elapsed:.1f}s")

        return m["loss"] / m["total"], m["correct"] / m["total"]

    @torch.no_grad()
    def validate(self):
        self.model.eval()
        loss_sum, correct, total = 0, 0, 0
        for i, batch in enumerate(self.val_loader):
            if i >= self.cfg.val_batches: break
            l, c, t = self._step(batch, is_train=False)
            loss_sum += l * t; correct += c; total += t
        return loss_sum / total, correct / total

    def fit(self):
        total_start = time.time()
        print(f"🚀 Старт обучения: {datetime.now().strftime('%H:%M:%S')}")
        print(f"Device: {self.cfg.device} | Voxel Size: {self.cfg.voxel_size}")

        for epoch in range(1, self.cfg.epochs + 1):
            t_loss, t_acc = self.run_epoch(epoch)
            v_loss, v_acc = self.validate()
            
            # Красивый итог эпохи
            is_best = v_acc > self.best_acc
            marker = "⭐ NEW BEST" if is_best else ""
            
            summary = (
                f"\n📊 ИТОГ ЭПОХИ {epoch}:\n"
                f"   🔹 Train Loss: {t_loss:.4f} | Acc: {t_acc:.2%}\n"
                f"   🔸 Val   Loss: {v_loss:.4f} | Acc: {v_acc:.2%}  {marker}\n"
                f"{'-'*50}"
            )
            print(summary)
            self.history.append(summary)
            
            if is_best:
                self.best_acc = v_acc
                self.save_checkpoint(epoch)
            
            torch.cuda.empty_cache()

        total_time = (time.time() - total_start) / 60
        print(f"\n🎉 Обучение завершено за {total_time:.1f} мин!")
        print(f"🏆 Лучшая точность (Val Acc): {self.best_acc:.2%}")
        self.save_logs()

    def save_checkpoint(self, epoch):
        torch.save({"state": self.model.state_dict(), "acc": self.best_acc}, self.cfg.weights)

    def save_logs(self):
        with open("training_report.log", "w", encoding="utf-8") as f:
            f.write("\n".join(self.history))
```
